```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

# Data Acquisition

## Read the CSV file into a DataFrame
tips = pd.read_csv('tips.csv')
```

```python
# Inspect first rows of the DataFrame
print(tips.columns)
```

```python
# Initial data review
print(tips.head(20))
print(tips.shape)
print(tips.dtypes)

# Stats run
print(tips.describe())
```

```python
# Inspect histogram by numerical column
tips['total_bill'].hist()

# Inspect histograms for all numerical columns
tips.hist()

# Show all plots
plt.show()
```

```python
# Inspect value counts by categorical column
print(tips['sex'].value_counts())  # Example for a categorical column

# Inspect value counts for all categorical columns
for col in tips.select_dtypes(include=['object']).columns:  # Selecting object (categorical) columns
    # Display count plot
    sns.countplot(x=col, data=tips)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

```python
# Inspect value counts by categorical column
print(tips['tip'].value_counts())  # Example for a categorical column

# Inspect value counts for all categorical columns
for col in tips.select_dtypes(include=['object']).columns:  # Selecting object (categorical) columns
    # Display count plot
    sns.countplot(x=col, data=tips)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=tips, x='tip', y='day', color='Red')

plt.title('Tips per Day')
plt.xlabel('tip')
plt.ylabel('day')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(data=tips, x='tip', y='M_F', color='purple')

plt.title('tips per Day')
plt.xlabel('tip')
plt.ylabel('M_F')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips.rename(columns={'total_bill': 'Bill', 'sex': 'M_F', 'size': 'Party_Size'}, inplace=True)

tips['Total Party_Size per day'] = tips.groupby('day')['Party_Size'].transform('sum')

tips['Percent Change in Bill'] = tips['Party_Size'].pct_change() * 100

tips['Average tip per day'] = tips['Party_Size'].mean()
tips['Bill and tip'] = tips['Bill'].astype(str) + '-' + tips['tip'].astype(str)

g = sns.FacetGrid(tips, col='day', hue='M_F')
g.map(sns.scatterplot, 'Bill', 'tip')
g.add_legend()

sns.pairplot(tips)
plt.show()
```

```python
## Observations
- Men received higher tips than women.
- Tips are higher on Sundays during dinner time.
- Data for smoker has no relevance on the amount of tip left.
```
